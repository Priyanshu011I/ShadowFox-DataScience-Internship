```python
# Check duplicate rows
df.duplicated().sum()
```

```python
# Remove duplicate rows
df.drop_duplicates(inplace=True)
```

```python
# Fill missing values
df["director"] = df["director"].fillna("Unknown")
df["cast"] = df["cast"].fillna("Unknown")
df["country"] = df["country"].fillna("Unknown")
df["rating"] = df["rating"].fillna("Not Rated")
```

```python
# Movies vs TV Shows (Count Plot) Code

plt.figure(figsize=(8,5))
sns.countplot(x='type', data=df, palette='Set2')
plt.title("Movies vs TV Shows on Netflix")
plt.xlabel("Content Type")
plt.ylabel("Count")
plt.show()
```

```python
# Movies vs TV Shows (Pie Chart) Code

plt.figure(figsize=(7,7))
df['type'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%',
    startangle=90
)
plt.title("Percentage of Movies and TV Shows")
plt.ylabel("")
plt.show()
```

```python
# Top 10 Countries Code

plt.figure(figsize=(12,6))
df['country'].value_counts().head(10).plot(
    kind='bar',
    color='skyblue'
)
plt.title("Top 10 Countries with Most Netflix Content")
plt.xlabel("Country")
plt.ylabel("Number of Titles")
plt.xticks(rotation=45)
plt.show()
```

```python
# Content Rating Distribution Code

plt.figure(figsize=(10,5))
sns.countplot(
    y='rating',
    data=df,
    order=df['rating'].value_counts().index
)
plt.title("Distribution of Content Ratings")
plt.show()
```

```python
# Content Released by Year Code

plt.figure(figsize=(12,6))
df['release_year'].value_counts().sort_index().plot()
plt.title("Content Released by Year")
plt.xlabel("Release Year")
plt.ylabel("Number of Titles")
plt.show()
```

```python
# Heatmap Code

numeric_df = df.select_dtypes(include=['int64','float64'])
plt.figure(figsize=(8,6))
sns.heatmap(
    numeric_df.corr(),
    annot=True,
    cmap='coolwarm'
)
plt.title("Correlation Heatmap")
plt.show()
```
